```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# membaca gambar
image = cv2.imread('/content/drives/monarch.jpg')

# mengubah warna dari BGR menjadi RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# Membentuk kembali gambar menjadi array piksel 2D dan 3 nilai warna (RGB)
pixel_vals = image.reshape((-1,3))

# Konversi ke tipe float
pixel_vals = np.float32(pixel_vals)
```

```python
#baris kode di bawah ini menentukan kriteria agar algoritma berhenti berjalan,
#yang akan terjadi adalah 100 iterasi dijalankan atau epsilon (yang merupakan akurasi yang diperlukan)
#menjadi 85%
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)

# kemudian melakukan k-means clustering dengan jumlah cluster yang didefinisikan sebagai 3
# juga pusat acak pada awalnya dipilih untuk pengelompokan k-means
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# mengubah data menjadi nilai 8-bit
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# membentuk kembali data ke dalam dimensi gambar asli
segmented_image = segmented_data.reshape((image.shape))

plt.imshow(segmented_image)
```
